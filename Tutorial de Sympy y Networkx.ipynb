# INTRODUCIR UNA MATRIZ

Para crear una matriz en SymPy, se utiliza $$\textbf{Matrix}([ \quad ])$$

Una matriz se construye proporcionando una lista de vectores fila que la componen.

Dentro de los corchetes irán las filas de cada matriz igual entre corchetes separando cada entrada por una coma, por ejemplo: $$[1 , -1]$$ esa es la primer fila.

Para introducir una segunda fila, después del corchete de la fila anterior se coloca una coma y se abren corchetes nuevamente para ingresar la segunda fila, por ejemplo: $$[1,-1] , [3, 4]$$

Aquí ya se tienen dos filas de la matriz.

Finalmente la forma completa sería

$$\textbf{Matrix}([ [1 , -1] , [3, 4] ])$$

Lo anterior imprime una matriz $2 $ x $ 2$, pero se puede poner una matriz de dimensión $n$ x $n$ con $n$ un número entero.

Por ejemplo:

```python
Matrix([[5, -1, 2], [9, -4, 3], [1, 2, 6]])
```

Para llamar a la Matriz anterior se le puede asignar una letra, con esto la nombramos y ya no es necesario escribir toda la matriz en casos posteriores, basta con únicamente llamarla por el nombre que se le puso.

Por ejemplo:

```python
A = Matrix([[5, -1, 2], [9, -4, 3], [1, 2, 6]])
```

Le asigne el nombre A, pero pudo haber sido cualquier otro, ahora basta con escribir su nombre en la línea de código para llamarla:

```python
A
```

# Multiplicación de matrices

Ahora para multiplicar matrices, basta con usar el operador "*" entre las matrices que se desea multiplicar, por ejemplo:

$$M \text{*} N$$

Con A una matriz de $n \text{x} m$ dimensiones y B una matriz de $m \text{x} p$ dimensiones. Las dimensiones son importantes para saber si es posible multiplicar matrices o vectores.

Ejemplo:

```python
B=Matrix([[2,  1], [-3,  7], [5, 0]])
```

Se crea una matriz N de $3 \text{x} 2$ dimensiones, que es compatible con la matriz A que se creo anteriormente

```python
B
```

Ahora se puede llevar acabo la multiplicación

```python
A*B
```

Se tiene el resultado de la multiplicación, que nuevamente es una matriz

#Extración de columna o fila de una matriz

De igual manera se puede obtener una fila, columna o entrada especifica de una matriz, esto de la forma

$$A[i,:]$$

Se obtiene la fila $i.$

```python
A[1,:]
```

Para la columna es de manera similar:

$$A[:,i]$$

Se obtiene la columna $i$.

```python
A[:,0]
```

Y para obtener una entrada especifica, se pone la posición en la que se encuentra:

$$A[i,j]$$

Se obtiene la entrada $i,j \text{ con }\quad i,j=0,1,2,...n-1$

```python
A[1,1]
```

#Elevar una a una potencia una matriz

Elevar una matriz a una potencia dada es de la forma:

$$A ^n \text{ con } n=0,1,2,...$$

```python
A**2
```

# $\color{blue}{\text{TUTORIAL DE NETWORKX PARA DIBUJAR DIAGRAMAS DE TRANSICIÓN}}$

Se importa la libreria networkx

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

Para crear un gráfico vacio es:

```python
G = nx.Graph()
```

El grafo G se puede hacer crecer de varias maneras. NetworkX incluye muchas funciones de generador de grafos y facilidades para leer y escribir grafos en muchos formatos. Sin embargo, para empezar, veremos manipulaciones sencillas. Puedes añadir un nodo a la vez,

```python
G.add_node(1)
```

o agregar nodos desde cualquier contenedor iterable , como una lista

```python
G.add_nodes_from([2, 3])
```

También puede agregar nodos junto con atributos de nodo si su contenedor produce 2 tuplas del formato

```python
G.add_nodes_from([(4, {"color": "red"}), (5, {"color": "green"})])
```

Los nodos de un gráfico se pueden incorporar a otro:

```python
H = nx.path_graph(10)
G.add_nodes_from(H)
```

G Ahora contiene los nodos de H como nodos de G. Por el contrario, podrías usar el gráfico H como un nodo en G.

```python
G.add_node(H)
```

El gráfico Gahora contiene Hun nodo. Esta flexibilidad es muy potente, ya que permite crear gráficos de gráficos, gráficos de archivos, gráficos de funciones y mucho más.

#Aristas

G también se puede aumentar añadiendo una arista a la vez,

```python
G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)  # unpack edge tuple*
```

añadiendo una lista de aristas,

```python
G.add_edges_from([(1, 2), (1, 3)])
```

Se pueden eliminar nodos y aristas del grafo de forma similar a como se agregan. Utilice los métodos Graph.remove_node(), Graph.remove_nodes_from(), Graph.remove_edge() y Graph.remove_edges_from().

Al eliminar un nodo también se eliminan sus bordes incidentes:

```python
G.remove_node(2)
G.remove_nodes_from("spam")
```

#Ejemplo

Para crear un gráfico dirigido es con:

DiGraph()

```python
# Crear grafo dirigido
G = nx.DiGraph()
```

Supongamos que tenemos 3 estados: A, B, C, y queremos indicar probabilidades de transición:

```python
# Agregar transiciones
G.add_edge("A", "B", weight=0.6)
G.add_edge("B", "C", weight=0.4)
G.add_edge("C", "A", weight=1.0)
```

Podemos dibujar usando nx.draw y mostrar las etiquetas de las aristas (transiciones):

```python
# Posición automática de los nodos
pos = nx.spring_layout(G)

# Dibujar grafo y etiquetas de las aristas
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G,'weight'))

plt.show()
```

Se puede mejorar el diagrama poniendo titulo, color a los nodos, aumentar el tamaño, de la misma manera en que se gráfica comunmente.

```python
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=1500, arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G,'weight'))
plt.title("Diagrama de Transición")
plt.show()
```

#Ejemplo usando matriz de transición

Un ejemplo muy simple.

Pero ahora, dada una matriz de transición se creará el diagrama

Se crea un arreglo numpy para facilitar el grafo

```python
P = np.array([[0, 0.6, 0.4],
              [0, 0.5, 0.5],
              [0.8, 0.2, 0]])
```

Se les da nombres a los estados

```python
# Nombres de los estados
estados = ["A", "B", "C"]
```

Se usa un comando de networkx y numpy para crear el diagrama a partir de la matriz

```python
G = nx.from_numpy_array(P, create_using=nx.DiGraph)  # crea el grafo desde la matriz
```

Finalmente se gráfica de la misma manera

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.show()
```

O bien, se pueden agregar de manera más manual incluyendo todo lo deseado, de la manera siguiente:

```python
# Crear grafo dirigido
G = nx.DiGraph()

# Agregar aristas con probabilidades
for i, fila in enumerate(P):
    for j, prob in enumerate(fila):
        if prob > 0:
            G.add_edge(estados[i], estados[j], probabilidades=prob)
```

Se gráfica de la misma manera, incluyendo ahora las probabilidades

```python
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True)

# Dibujar etiquetas de las aristas con las probabilidades
edge_labels = nx.get_edge_attributes(G, 'probabilidades')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.show()
```
